```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, Point
import numpy as np
import pandas as pd
```

```python
p = Polygon([(0, 0), (66.4, 0), (66.4, 11.2), (17.1, 11.2), (17.1, 16.6), (24, 16.6), (24, 22.9), (17.1,22.9), (17.1,43), (37.2,43), (37.2,54), (0,54)])

x,y = p.exterior.xy
```

```python
fig = plt.figure(1, figsize=(5,5), dpi=90)
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_title('Polygon Edges')
```

```python
df = pd.DataFrame.from_dict({'x' : [10.9,0.4,32.7,12.0,7.7,13.0,14.8,26.4],
                             'y' : [0.6,19.9,0.2,32.5,1.0,18.0,4.9,4.2]})
x_2 = df['x'].values
y_2 = df['y'].values

u = np.diff(x_2)
v = np.diff(y_2)
pos_x = x_2[:-1] + u/2
pos_y = y_2[:-1] + v/2
norm = np.sqrt(u**2+v**2) 

fig = plt.figure(1, figsize=(5,5), dpi=90)
ax = fig.add_subplot(111)
ax.plot(x,y)
ax.plot(x_2,y_2, marker="o", color="red")
ax.quiver(pos_x, pos_y, u/norm, v/norm, angles="xy", zorder=5, pivot="mid")
ax.set_title(""" Design Phase
Trajectory made by "data:TrackingTag_e9b870d9-5058-4d4a-802b-3457279524e0"(Tracked Data) in "data:Element_151035"(IFC Data)
""")
```
